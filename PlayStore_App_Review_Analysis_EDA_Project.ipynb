### Variables Description

* App: The App names

   been downloaded and installed on Android devices.
* Type:Type means weather the app is free or paid.

* Content Rating:"Content Rating" refers to a label assigned to apps and games
  that indicates the age appropriateness of their content. It helps users, particularly parents and guardians, to make informed decisions about which apps and games are suitable for different age groups.
* Genres: "Genres" refer to the different categories or classifications that apps and games are organized into based on their content, theme, or functionality. Genres are used to group similar apps together, making it easier for users to discover and explore apps that align with their interests or needs.
* Last Updated: Date given on which you last updated your app.
* Current Ver:, "Current Version" refers to the most recent release or update of a particular mobile application or game available for download. It is the version of the app that developers have made available to users at the time of visiting the app's page on the Play Store.
* Android Ver:"Android version" refers to the minimum or recommended version of the Android operating system required to run a particular mobile application or game.

````python
# to fill column type,current ver,android ver with mode
playstore_df['Type'].fillna(str(playstore_df['Type'].mode().values[0]),inplace=True)
playstore_df['Current Ver'].fillna(str(playstore_df['Current Ver'].mode().values[0]),inplace=True)
playstore_df['Android Ver'].fillna(str(playstore_df['Android Ver'].mode().values[0]),inplace=True)
````

````python
playstore_df.describe()
````

````python
playstore_df.head()
````

````python
playstore_df['Price']=playstore_df['Price'].apply((lambda x: str(x).replace('$', '') if '$' in str(x) else str(x)))
# applying function through lambda and replacing $ value with space else dont change anything
playstore_df['Price']=playstore_df['Price'].apply(lambda x : float(x))
# converting all price values to float
playstore_df['Reviews']=pd.to_numeric(playstore_df['Reviews'],errors='coerce')
````

````python
playstore_df['Installs']=playstore_df['Installs'].apply(lambda x : str(x).replace('+','') if '+' in str(x) else str(x))
playstore_df['Installs']=playstore_df['Installs'].apply(lambda x : str(x).replace(',','') if ',' in str(x) else str(x))
playstore_df['Installs']=playstore_df['Installs'].apply(lambda x: float(x))
````
